```python
#TO INCRESAE THE SIZE OF POINTS
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
size= np.array([100,200,300,400])
plt.scatter(x,y, s = size , alpha=0.5)
plt.show()
```

```python
#TO INCRESAE the SIZE , COLOR 
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
size= np.array([100,200,300,400])
c= np.array(['red', 'green', 'blue','yellow'])
plt.scatter(x,y, s = size ,c=c, alpha=0.5) # alpha use to show the points overlapped by others
plt.show()
```

```python
x= np.array([3,5,8,6,9,12])
y= np.array([12,34,56,18,45,78])
sizes= np.array([20,10,30,40,50,60])
c=np.array(['red', 'black' , 'pink', 'orange', 'green', 'yellow'])
plt.scatter(x,y,s=sizes, alpha=0.5, c=c, cmap="nipy_spectral") # use to print the s is used for size  
                                                                #alpha use for viewing, c for color , cmap for color bar 
plt.colorbar()
plt.show()
```

### barplot

```python
x=np.array(['A', 'B', 'C', 'D']) # use to print the bar plot
y= np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(['A', 'B', 'C', 'D']) # use to print the bar plot
y= np.array([3,8,1,10])
plt.barh(x,y)  # barh is use to show the horizontal bar
plt.show()
```

```python
x=np.array(['A', 'B', 'C', 'D']) # use to draw the bar 
y= np.array([3,8,1,10])
plt.barh(x,y, color= 'red') # use to show the red color of bar lines 
plt.show()
```

```python
x=np.array(['A', 'B', 'C', 'D'])  # use to print barplot 
y= np.array([3,8,1,10])
plt.bar(x,y, color= 'red', width= 0.1) # width to decrease the bar line
plt.show()
```

# PIECHART

```python
y= np.array([1,2,3,4]) # use to dram the pie chart
plt.pie(y)
plt.show()
```

```python
y= np.array([1,2,3,4]) # use to draw the pie chart
mylabel= ["apple" , "mango", "banana", "oragne"]
plt.pie(y , labels=mylabel)  # label use to label the pie chart 
plt.show()
```

```python
y= np.array([1,2,3,4]) # use to draw the pie chart
mylabel= ["apple" , "mango", "banana", "oragne"]
plt.pie(y , labels=mylabel, startangle=90)  # label use to label the pie chart 
plt.show()
```

```python
y= np.array([1,2,3,4]) # use to draw the pie chart
mylabel= ["apple" , "mango", "banana", "oragne"]
myexplode= [0.2,0,0,0]
plt.pie(y , labels=mylabel, startangle=90, explode=myexplode)  # use for drag it apple 
plt.show()
```

```python
y= np.array([1,2,3,4]) # use to draw the pie chart
mylabel= ["apple" , "mango", "banana", "oragne"]
myexplode= [0.2,0,0,0]
plt.pie(y , labels=mylabel, startangle=90, explode=myexplode , shadow= True)  # use for shadow 
plt.show()
```

```python
y= np.array([1,2,3,4]) # use to draw the pie chart
mylabel= ["apple" , "mango", "banana", "oragne"]
myexplode= [0.2,0,0,0]
mycolor=["red", "orange", "black", "yellow"]
plt.pie(y , labels=mylabel, startangle=90, explode=myexplode, shadow=True, colors=mycolor, autopct='%1.1f%%')  # use for drag it apple 
plt.show()
```

```python
y= np.array([1,2,3,4]) # use to dram the pie chart
mylabel = ["apple", "mango", "banana", "orange"] #to show the label of the pie chart
myexplode = [0.2,0,0,0]
mycolor = ['red','yellow','blue','cyan']
plt.pie(y, labels = mylabel, startangle = 90, explode= myexplode, shadow = True, colors= mycolor, autopct= '%1.1f%%')# give shadow to the pie chart
plt.legend(title = "color")
plt.show()
```
